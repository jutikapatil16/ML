```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as py
```

```python
x = [1,2,3,4]
y=[11,22,33,44]

# setup
fg, ax = plt.subplots(figsize= (10,10))

# plot
ax.plot(x,y)

# customize
ax.set(title="title", xlabel="x", ylabel="y")
```

```python
x = np.linspace(0,10,10)


# 1. line plot
# setup
fg, ax = plt.subplots()

# plot
ax.plot(x,x**2)
```

```python
x = np.linspace(0,10,100)

# 2. Scatter
# setup
fg, ax = plt.subplots()

# plot
ax.scatter(x,np.exp(x))
```

```python
nut_butter_price = {"Almond butter":10,
                   "Peanut butter":8,
                   "Cashew butter": 6}
fg, ax = plt.subplots()
ax.bar(nut_butter_price.keys(),nut_butter_price.values());
ax.set(title="Dan's nut butter store" ,xlabel ="butter", ylabel="Price$")
```

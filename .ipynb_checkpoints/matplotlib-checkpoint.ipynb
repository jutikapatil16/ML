```python
nut_butter_price = {"Almond butter":10,
                   "Peanut butter":8,
                   "Cashew butter": 6}
fg, ax = plt.subplots()
ax.bar(nut_butter_price.keys(),nut_butter_price.values());
ax.set(title="Dan's nut butter store" ,xlabel ="butter", ylabel="Price$")
```

```python
fg, ax = plt.subplots()
ax.barh(list(nut_butter_price.keys()),list(nut_butter_price.values()));
ax.set(title="Dan's nut butter store" ,ylabel ="butter", xlabel="Price$")
```

```python
fig ,((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows = 2, ncols =2, figsize=(10,5))

x = np.random.randn(100)
y = np.linspace(11,20,num=100)

ax1.plot(x,x/2)
ax2.scatter(x*x,np.random.randn(100))
ax3.bar(x,y)
ax4.hist(x)
```

```python
# same thing other way
fig ,ax = plt.subplots(nrows = 2, ncols =2, figsize=(10,5))

x = np.random.randn(100)
y = np.linspace(11,20,num=100)

ax[0,0].plot(x,x/2)
ax[0,1].scatter(x*x,np.random.randn(100))
ax[1,0].bar(x,y)
ax[1,1].hist(x)
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]','')
car_sales

# type is str therefore str
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales

# cumcum gives cumulative frequency
# astype necause its a string, it will result in  400000500000
```

```python
car_sales.plot(x= "Total Sales",y="Price")

# as Price is not numeric 
```

```python
car_sales["Price"] = car_sales["Price"].astype(int)
car_sales.plot(x= "Total Sales",y="Price")
```

```python
car_sales.plot(x= "Total Sales",y="Price", kind ="scatter")

# when csv se read then x = " name of the colum"
```

```python
heart_disease =pd.read_csv("heart-disease.csv")
heart_disease.head()
```


# TWO WAYS OO METHOD AND PyPlot

```python
over_50 = heart_disease[heart_disease["age"] >50]
over_50
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="sex" , figsize=(10,10))
```

```python
# oo
fix , ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter( x=over_50["age"], y=over_50["chol"], c=over_50["sex"])

# c for determining which is 1 and which is 0

ax.set_xlim([40,70])

ax.set(title="Heart  Disease and Cholestrol Levels", 
       xlabel="Age",
       ylabel="Cholestrol")

ax.legend(*scatter.legend_elements(), title="Sex");


ax.axhline(over_50["chol"].mean(), linestyle="-.")
```
